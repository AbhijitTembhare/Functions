## FUNCTIONS

 ### **1. What is the difference between a function and a method in Python?**  
- A **function** is an independent block of reusable code.  
- A **method** is a function associated with an object.  
- **Example:**  
  ```python
  # Function
  def add(a, b):
      return a + b
  print(add(2, 3))  # Output: 5  

  # Method
  text = "hello"
  print(text.upper())  # Output: "HELLO"
  ```

### **2. Explain the concept of function arguments and parameters in Python.**  
- **Parameters** are placeholders in function definitions.  
- **Arguments** are actual values passed when calling a function.  
- **Example:**  
  ```python
  def greet(name):  # 'name' is a parameter
      return "Hello, " + name
  print(greet("Alice"))  # Output: "Hello, Alice"
  ```

### **3. What are the different ways to define and call a function in Python?**  
- **Normal function:**  
  ```python
  def add(a, b):
      return a + b
  print(add(2, 3))  # Output: 5
  ```  
- **Lambda function:**  
  ```python
  add = lambda a, b: a + b
  print(add(2, 3))  # Output: 5
  ```  
- **Default arguments:**  
  ```python
  def greet(name="User"):
      return "Hello " + name
  print(greet())  # Output: "Hello User"
  ```  
- **Arbitrary arguments (`*args`):**  
  ```python
  def sum_all(*numbers):
      return sum(numbers)
  print(sum_all(1, 2, 3, 4))  # Output: 10
  ```

### **4. What is the purpose of the `return` statement in a Python function?**  
- The `return` statement sends back a result from a function.  
- **Example:**  
  ```python
  def square(x):
      return x * x
  print(square(4))  # Output: 16
  ```

### **5. What are iterators in Python and how do they differ from iterables?**  
- **Iterable:** An object that can be looped over (`list`, `tuple`, `dict`).  
- **Iterator:** An object that produces values using `next()`.  
- **Example:**  
  ```python
  my_list = [1, 2, 3]
  iterator = iter(my_list)  # Creating an iterator
  print(next(iterator))  # Output: 1
  print(next(iterator))  # Output: 2
  ```

### **6. Explain the concept of generators in Python and how they are defined.**  
- **Generators** use the `yield` keyword to produce values lazily.  
- **Example:**  
  ```python
  def gen_nums():
      yield 1
      yield 2
  gen = gen_nums()
  print(next(gen))  # Output: 1
  print(next(gen))  # Output: 2
  ```

### **7. What are the advantages of using generators over regular functions?**  
- **Memory-efficient** (does not store all values at once).  
- **Lazy evaluation** (produces values when needed).  
- **Example:**  
  ```python
  def infinite_numbers():
      num = 0
      while True:
          yield num
          num += 1
  gen = infinite_numbers()
  print(next(gen))  # Output: 0
  print(next(gen))  # Output: 1
  ```

### **8. What is a lambda function in Python and when is it typically used?**  
- A **lambda function** is a small anonymous function.  
- Typically used in `map()`, `filter()`, or as a quick inline function.  
- **Example:**  
  ```python
  square = lambda x: x * x
  print(square(5))  # Output: 25
  ```

### **9. Explain the purpose and usage of the `map()` function in Python.**  
- `map()` applies a function to all items in an iterable.  
- **Example:**  
  ```python
  numbers = [1, 2, 3]
  squared = list(map(lambda x: x * x, numbers))
  print(squared)  # Output: [1, 4, 9]
  ```

### **10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**  
- **`map()`**: Applies a function to all elements.  
- **`filter()`**: Filters elements based on a condition.  
- **`reduce()`**: Performs a cumulative operation.  
- **Example:**  
  ```python
  from functools import reduce
  numbers = [1, 2, 3, 4]

  # Using map
  doubled = list(map(lambda x: x * 2, numbers))
  print(doubled)  # Output: [2, 4, 6, 8]

  # Using filter
  even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
  print(even_numbers)  # Output: [2, 4]

  # Using reduce
  sum_all = reduce(lambda x, y: x + y, numbers)
  print(sum_all)  # Output: 10
  ```

### **11. Internal Mechanism for Sum Operation (`sum([47,11,42,13])`)**  
- The `sum()` function internally uses an iterative approach.  
- **Implementation:**  
  ```python
  total = 0
  for num in [47, 11, 42, 13]:
      total += num
  print(total)  # Output: 113
  ```  
- **Step-by-step breakdown:**  
  ```
  Initial total = 0  
  Add 47 → total = 47  
  Add 11 → total = 58  
  Add 42 → total = 100  
  Add 13 → total = 113  
  ```
- **Final Output:** `113`
